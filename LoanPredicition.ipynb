```python
#we drop loan_id and no_of dependents as it is not relevent as evident from the correlation
data.drop(['loan_id',' no_of_dependents'],axis=1,inplace=True)
```

```python
#we need to apply standard scalar to the numeric data
from sklearn.preprocessing import StandardScaler
import numpy as np
numeric_data=data.select_dtypes(include=[np.number])
categorical_data=data.select_dtypes(exclude=[np.number])
```

```python
numeric_data
```

```python
categorical_data
```

```python
#Now that we have speparated numeric data we apply standard scalar to it
sc=StandardScaler()
scaled_numeric_data=sc.fit_transform(numeric_data)
```

```python
scaled_numeric_data
```

```python
#We convert this to pd dataframe
scaled_numeric_data_df=pd.DataFrame(scaled_numeric_data,index=data.index,columns=numeric_data.columns)
scaled_numeric_data_df
```

```python
remaining_df=numeric_data[[' income_annum',' loan_amount',' loan_term',' cibil_score']]
```

```python
 #We now apply PCA on residential_assets_vale ,commerical_asset_value,luxury_assets_value and bank_asset_value
#to reduce the dimensions of the data
from sklearn.decomposition import PCA
assets_pca=scaled_numeric_data_df[[' residential_assets_value',' commercial_assets_value',' luxury_assets_value',' bank_asset_value']]
Pca=PCA(n_components=1)
Compressed_assets=Pca.fit_transform(assets_pca)
```

```python
Compressed_assets
```

```python
Compressed_assets_pd=pd.DataFrame(Compressed_assets,index=data.index,columns=['Total Assets'])
```

```python
#Now we use get dummies to convert the categorical 0s and 1s
categorical_data_dummies=pd.get_dummies(categorical_data,drop_first=True)
```

```python
#finally we merge the datasets
Final_data=pd.concat([categorical_data_dummies,remaining_df,Compressed_assets_pd],axis=1)
```

```python
Final_data
```

```python
X=Final_data.drop([' loan_status_ Rejected'],axis=1)
y=Final_data[' loan_status_ Rejected']
```

```python
#TEST AND TRAIN SPLIT
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier 
Dc=DecisionTreeClassifier()
Dc.fit(X_train,y_train)
y_pred=Dc.predict(X_test)
```

```python
from sklearn import metrics
from sklearn.metrics import confusion_matrix
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
confusion_matrix(y_test,y_pred)
```
